```python
import seaborn as sns 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('diabetes.csv')
pf = df.head(10)
pf
```

```python
sns.distplot(df['Glucose'])
```

```python
sns.jointplot(x ='BloodPressure', y ='SkinThickness', data= df, kind = 'kde')
```

```python
sns.kdeplot(df['Insulin'], shade = True, bw=20)
```

```python
sns.kdeplot(df['Insulin'],df['Age'], shade_lowest = False, shade = True, cbar = True)
```

```python
sns.pairplot(df)
```

```python
sns.boxplot(x='Glucose', y ='Age', data = pf)
```

```python
plt.figure(figsize=(12,8))
sns.stripplot(x='Outcome', y='Insulin', data= df, hue = 'Pregnancies', dodge = True)
```

```python
df_mx=df.corr()
df_mx
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df_mx, cmap='Blues')
```

```python
sns.clustermap(df.head(25), standard_scale= 1)
```

```python
df_fg = sns.FacetGrid(df, hue = 'Outcome', row = 'Pregnancies')
df_fg.map(plt.scatter, 'BMI', 'Insulin')
```
